```julia
using PyCall, PyPlot
```

```julia
@pyimport numpy.random as random
@pyimport pandas as pd
@pyimport seaborn as sns

x = random.normal(size=100)

titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

```julia
sns.distplot(x, kde=false, rug=false, bins=10) 
```

```julia
sns.jointplot("sepal_width", "petal_length", data=iris)
```

```julia
sns.pairplot(iris, hue="species")
```

```julia
sns.stripplot(x="day", y="total_bill", data=tips)
```

```julia
sns.stripplot(x="day", y="total_bill", data=tips, hue="time")
```

```julia
sns.boxplot(x="size", y="tip", data=tips[:sort]("size"))
```

```julia
sns.barplot(x="sex", y="survived", data=titanic, hue="class")
```

```julia
sns.countplot(x="sex", hue="embarked", data=titanic, palette="Greens_d")
```

```julia
df_flights = sns.load_dataset("flights")
df_flights[:head](5)
```

```julia
grid = sns.FacetGrid(df_flights, col="year", hue="year", col_wrap=4, size=5)
grid[:map](sns.pointplot, "month", "passengers")
```

```julia
df_flights_pivot = pd.pivot_table(data=df_flights, values="passengers", 
                                  columns="year", index="month", aggfunc=mean)
```

```julia
sns.heatmap(df_flights_pivot)
```
